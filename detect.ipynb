```python
import yolov5
import os
import random
import matplotlib.pyplot as plt
import cv2 as cv
```

## Load custom model with ".pt" file

```python
class_names={
    0:"-0",
    1:"-1",
    2:"adult",
    3:"kid"
    }

model = yolov5.load('data/weights/best.pt', device='cpu')
# set model parameters
model.conf = 0.25  # NMS confidence threshold
model.iou = 0.45  # NMS IoU threshold
model.agnostic = False  # NMS class-agnostic
model.multi_label = False  # NMS multiple labels per box
model.max_det = 1000  # maximum number of detections per image
model.names = class_names
```

## Load dataset

```python
# Load image
test_images_path = os.path.join("data/dataset/train")
test_images_names = os.listdir(test_images_path)
```

## Test randomly the model

```python
image_path = os.path.join(test_images_path,random.choice(test_images_names)) 

results = model("data/images/test10.PNG")
results.show()
```
